```python
%matplotlib inline
import cv2
import matplotlib
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread('/Users/VanessaG/Desktop/lunch.jpg')
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def show(image):
    plt.figure(figsize=(8, 8))
    plt.imshow(image, interpolation='nearest')
    plt.xticks([], [])
    plt.yticks([], [])

show(image)
```

```python
images = []
for i in [0, 1, 2]:
    color = image.copy()
    if i != 0: color[:,:,0] = 0
    if i != 1: color[:,:,1] = 0
    if i != 2: color[:,:,2] = 0
    images.append(color)

show(np.hstack(images))
```
